**1. Explain the Activation Functions in your own language  
a)	sigmoid  
b)	tanh  
c)	ReLU  
d)	ELU  
e)	LeakyReLU  
f)	swish**  


- Ans : 
     a) sigmoid - Sigmoid activation function is defined as : $\sigma(x) = \frac{e^x}{1 + e^x}$. The output of this function ranges between 0 and 1. The derivative of this function is given as $\frac{d}{dx}(\sigma(x)) = \sigma(x). (1 - \sigma(x))$. The main drawback of this function is that the function value saturates either at tail of zero or 1. 
     The derivarive of this function becomes very small at these regions. In this case, the near-zero derivative would make the gradient of the loss function very small, which prevents the update to the weights and thus the entire learning process. Also, this function is not zero centered which makes the neural network difficult and unstable.
     
     
         
     b)	tanh - Tanh activation function is defined as : $tanh = \frac{e^x - e^{-x}}{e^x + e^{-x}}$. The output of this function ranges between -1 and +1. The derivative of this function is given as : $\frac{d}{dx}(tanh(x)) = 1 - tanh^2(x)$. 
     Unlike sigmoid, the values and derivatives for tanh function is zero centered. The gradient of tanh function is steeper than that of sigmoid. Also in this case, the gradient are not restricted to move in a certain direction. 
     
     
     
     c)	ReLU -   ReLu is known as Rectified Linear Unit and is defined as $R(x) = x$ if $x \ge 0$ and $R(x) = 0$ otherwise. The derivative of this function becomes : $\frac{d}{dx} (R(x)) = 1 $ if $x \ge 0$, 0 otherwise. 
    The functional form of ReLU is simpler than the sigmoid and tanh functions, i.e. this function does not have exponential dependence and hence is faster. Also, because of the linear and non saturating property of ReLu, the the convergence of gradient descent towards the global minimum of the loss function is faster in comparison to other activation functions. Along with the advantages of ReLu, it has one big disadvantage known as 'dying ReLu' problem. This occurs during the weight updation. If the weight of any neuron becomes negative then the output is zero for these neurons. This means that the hidden values of these neurons are always zero and do not contribute to the training process. This means that the gradient flowing through these ReLU neurons will also be zero from that point on. These are known as 'dead neurons'. It has been found that 20-50 percent of the entire neural network that used ReLU activation can be dead.    
    
    
     d)	ELU - ELU is known as Exponential Linear Unit and is defined as : $ELU (x) = x$ if $x \ge 0$ and $ELU (x) = \alpha (e^x -1)$ if $x \lt 0$. This function is very similar to ReLU except the negative inputs. The derivative of this function is given as : $\frac{d}{dx} (ELU(x)) = 1$ if $x \ge 0$ and $\frac{d}{dx} (ELU(x)) = \alpha . e^x$ if $x \lt 0$ 
     This function smoothens out slowly until its output equals to $- \alpha$. This is very good alternative of ReLU that helps solving the 'dead neuron' problem. But this function also has a problem. For the range $x \gt 0$, this activation function can blow up in range [$0, \infty$].
     
     
     e)	LeakyReLU - LeakyReLu is a very good alternative of ReLU, which takes care of the 'dead neuron' problem. This function is defined as :$f(x) = x $ if $x \ge 0$ and $f(x) = \alpha x$ if $x \lt 0$. The derivative of this function is given as : $\frac{d}{dx} (f(x)) = 1$ if $x \ge 0$ and $\frac{d}{dx} (f(x)) = \alpha $ if $x \lt 0$ 
     The slope parameter $\alpha$ helps to overcome the 'dying ReLU' problem. Apart from the benefits, this activation function cannot be used for complicated problems due to its linearity. 
     
     f)	swish - Swish activation function was developed by Google. Swish is as computationally efficient as ReLU and shows better performance than ReLU on deeper models.  The values for swish ranges from negative infinity to infinity. The function is defined as : $f(x) = \frac{1}{1 + e^{-x}} = x . \sigma(x)$. The derivative is given as : $\frac{d}{dx}(f(x)) = \sigma(x) + x . \frac{d}{dx} (\sigma(x))$. 
     A unique fact about this function is that swich function is not monotonic. This means that the value of the function may decrease even when the input values are increasing. 
     
     
     

**2. What happens when you increase or decrease the optimizer learning rate?** 

- Ans : If the learning rate is small, the model needs more epoch for training, resulting in larger time for convergence which can also result the learning process to get stuck. This increases the computing cost. Higher learning rate results in rapid changes and require less epochs. This results in faster convergence. Many a times, higher learning rate does not give best accuracy as the model does not get enough time to learn the data. 



**3. What happens when you increase the number of internal hidden neurons?** 


- Ans : Increasing the number of internal hidden neurons can help the model to capture the features better and improve the predictions. However, if the number of neurons are too much, this can lead to overfitting of model. As a result the model will describe the training data very nicely but will fail to generalize for the unseen data.

  


**4. What happens when you increase the size of batch computation?** 

- Ans : Batch size is one of the most important hyperparameters in deep learning training, and it represents the number of samples used in one forward and backward pass through the network and has a direct impact on the accuracy and computational efficiency of the training process. The batch size can be understood as a trade-off between accuracy and speed. Large batch sizes can lead to faster training times but may result in lower accuracy and overfitting, while smaller batch sizes can provide better accuracy, but can be computationally expensive and time-consuming.

  The batch size can also affect the convergence of the model, meaning that it can influence the optimization process and the speed at which the model learns. Small batch sizes can be more susceptible to random fluctuations in the training data, while larger batch sizes are more resistant to these fluctuations but may converge more slowly.

  It is important to note that there is no one-size-fits-all answer when it comes to choosing a batch size, as the ideal size will depend on several factors, including the size of the training dataset, the complexity of the model, and the computational resources available.



**5. Why we adopt regularization to avoid overfitting?**

- Ans : Regularization is one of the simplest and effective ways to prevent overfitting. This method adds a penalty term to the loss function, discouraging the model from assigning too much importance to any one feature. Two common forms of regularization are L1 and L2 regularization.

  L1 regularization adds the absolute values of the weights to the loss function. This encourages some weights to become exactly zero, effectively performing feature selection. It’s a handy tool when one suspects that only a subset of the features is essential.
  
   L2 regularization, on the other hand, adds the square of the weights to the loss function. This tends to evenly distribute the importance across all features, reducing the magnitude of weights and preventing them from growing too large.



**6. What are loss and cost functions in deep learning?** 

- Ans : Loss function calculates the difference in the actual and predicted values for a single sample of data. Cost function quantifies the overall cost or error of the model on the entire training set. 

  There are different kind of loss functions used, depending on the nature of the problem. 
  
  - Loss Function for Regression Problem : 
     - <u> Mean Absolute Error (MAE) </u>: MAE calculates the average of the absolute difference between the actual and predicted value : $MAE = \frac{1}{N} \sum_i^N |{Y_i}_{actual} - {Y_i}_{Pred} |$
     
     
     - <u> Mean Squared Error (MSE) </u> : MSE is calculated by taking average squared difference between the predictions and expected results. $MSE = \frac{1}{N} \sum_i^N ({Y_i}_{actual} - {Y_i}_{Pred} )^2$
     
     
     - <u> Root Mean Squared Error (RMSE) </u> : Root Mean Square error is the extension of MSE — measured as the average of square root of sum of squared differences between predictions and actual observations. $RMSE = \sqrt{\frac{1}{N} \sum_i^N ({Y_i}_{actual} - {Y_i}_{Pred} )^2 }$ 
  
  
  - Loss Function for Classification Problem : 
  
      - <u> Binary Cross Entropy </u> : This is used for binary classification problems. The loss function is also known as Log Loss. Binary cross entropy compares each of the predicted probabilities to the actual class output which can be either 0 or 1. It then calculates the score that penalizes the probabilities based on the distance from the expected value. That means how close or far from the actual value. $Log loss = - \frac{1}{N} \sum_i^N [{Y_i}_{actual} log ({Y_i}_{Pred}) + (1 - {Y_i}_{actual} log (1 - {Y_i}_{Pred}))]$
      
      
      
      - <u> Categorical Cross Entropy </u> : Categorical Cross Entropy is used for multiclass classification or softmax regression. The loss function is given as : $Loss = - \sum_j^K {Y_j}_{actual} log {Y_j}_{Pred} $, where $K$ is the number of class data. The cost function becomes : $Cost = \frac{1}{n} \sum_i^n \sum_j^K {Y_{ij}}_{actual} log {Y_{ij}}_{Pred}$
      
      
      
      
      - <u> Sparse Categorical Cross Entropy </u> : Sparse Categorical Cross Entropy is a loss function that is commonly used in machine learning algorithms to train classification models. It is an extension of the Cross Entropy loss function that is used for binary classification problems. The sparse categorical cross-entropy is used in cases where the output labels are represented in a sparse matrix format. 
        If target column has One hot encode to classes like 0 0 1, 0 1 0, 1 0 0 then use categorical cross-entropy. and if the target column has Numerical encoding to classes like 1,2,3,4….n then use sparse categorical cross-entropy.

**7. What do you mean by underfitting in neural networks?**

- Ans : Underfitting means when the model is too simple that is not capable to learn the features and correlation among them for the training dataset. As a result training accuracy is very poor. This model should not be used for the unseen data. One way to improve the situation is to adding more data (if possible), make the model a bit more complicated by adding additional nodes or additional hidden layers or initializing some different activation functions etc. 



**8. Why we use Dropout in Neural Networks?** 

- Ans : Dropout is a popular regularization technique specifically designed for neural networks. During training, dropout randomly deactivates a fraction of neurons in each layer. This simulates training multiple neural networks in parallel, forcing the model to be more robust and preventing it from relying too heavily on any one neuron.
