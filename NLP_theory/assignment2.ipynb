**3.	What are Unigrams, Bigrams, Trigrams?** 

- Ans : In Natural Language Processing n-gram is a continuous sequence of n items generated from a given sample of text where the items can be characters or words and n can be any numbers like 1,2,3, etc.

  An n-gram of size 1 is referred to as a unigram size 2 is a bigram size 3 is a trigram. When N>3 this is usually referred to as four grams or five grams and so on.



**4.	How to generate n-grams from text?**


```python
import nltk
from nltk.util import ngrams

text1 = 'This is a lovely day'

n=[1,2,3]

for i in n: 
    Ngram = ngrams(text1.split(), int(i))
    if i == 1: 
        print('--------Unigram--------') 
        for grams in Ngram: 
            print(grams)
    elif i == 2:   
        print('--------Bigram--------') 
        for grams in Ngram: 
            print(grams)
    else : 
        print('--------Trigram--------') 
        for grams in Ngram: 
            print(grams)
```

**5.	Explain Lemmatization**

- Ans : Lemmatization is a text pre-processing technique used in natural language processing (NLP) models to break a word down to its root meaning to identify similarities. For example, a lemmatization algorithm would reduce the word better to its root word, or lemme, good.  In lemmatization, the final word does not get chopped off in arbitrary manner, in fact the lemma (word after lemmatization) does have a proper meaning. 


**6.	Explain Stemming**

- Ans : Stemming is a natural language processing technique that is used to reduce words to their base form, also known as the root form. It is an important step in text pre-processing, and it is commonly used in information retrieval and text mining applications. However, in stemming process many a times the words are chopped in a arbitrary manner. This reduces word in stem that is not an actual meaningful word.

**9.	Explain Noun Phrase (NP) chunking** 

- Ans : Noun chunk parsing is a process in natural language processing (NLP) where a sentence is analyzed to identify and extract the noun phrases or chunks. A noun chunk typically consists of a noun and its modifiers, which can include adjectives, determiners, prepositions, and other words that describe or specify the noun.

  The goal of noun chunk parsing is to break down a sentence into its constituent parts to extract meaningful information, particularly focusing on the nouns and their associated descriptors. This process is essential for various NLP tasks, including information extraction, text summarization, and understanding the structure of sentences.

  Let us consider an example: Sentence: "The large brown dog chased the cat."

  Noun chunks: 
   "The large brown dog": This is a noun phrase that consists of the determiner "the," followed by adjectives "large" and "brown," modifying the noun "dog."  
   "the cat": This is another noun phrase where "the" is the determiner for the noun "cat."  
    Noun chunk parsing involves using parsing techniques and grammatical rules to identify these chunks within a sentence. Techniques like part-of-speech tagging and syntactic parsing help in identifying the words' roles and relationships within the sentence, assisting in the extraction of meaningful noun phrases.

  For instance, in a syntactic tree representation, noun chunks might be identified by identifying subtrees that contain a noun as the head along with its modifiers.

   Noun chunk parsing is fundamental in many NLP applications where understanding the structure of sentences and extracting relevant information is crucial for further analysis and processing.

**10.	Explain Named Entity Recognition** 

- Ans : Named Entity Recognition, a crucial task in natural language processing (NLP) that involves identifying and categorizing named entities within a text into predefined categories such as names of persons, organizations, locations, expressions of times, quantities, monetary values, percentages, and more.

  The primary goal of NER is to recognize and classify entities within a body of text to extract meaningful information, understand relationships, and enable further analysis. Here's a breakdown of its key components:

   Identification of Entities: NER aims to identify specific entities or elements in a text. For instance, in the sentence "Apple Inc. was founded by Steve Jobs in Cupertino," NER would recognize "Apple Inc." as an organization, "Steve Jobs" as a person, and "Cupertino" as a location.

  Classification into Categories: Once identified, these entities are categorized into predefined classes or types. For example:

  Person: Names of individuals, such as "Steve Jobs."   
  Organization: Names of companies or groups, like "Apple Inc."  
  Location: Names of places, cities, countries, etc., such as "Cupertino."  
  Date/Time: References to specific dates or times, like "January 1, 2023." 
  Quantity: Expressions of measurements or quantities, such as "10 kilograms."  
  Others: Including monetary values, percentages, and more.  
  Contextual Understanding: NER often relies on the context of words in a sentence to accurately recognize and classify named entities. For example, "Apple" could refer to the organization "Apple Inc." in one context and the fruit in another.

  NER employs various techniques, including machine learning algorithms, deep learning models (such as recurrent neural networks, transformers like BERT, or CRF - Conditional Random Fields), and rule-based systems, often trained on annotated datasets where human annotators have labeled the named entities.

  NER has broad applications in information retrieval, question answering systems, content categorization, sentiment analysis, and more. It's a foundational step in understanding unstructured text data and extracting structured information, making it easier for machines to comprehend and process human language.
