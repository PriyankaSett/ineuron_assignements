**4. In supervised learning, what are classification and regression?** 

- Ans : Fundamentally, classification is about predicting a label and regression is about predicting a quantity. That classification is the problem of predicting a discrete class label output for an example. That regression is the problem of predicting a continuous quantity output for an example.

    This can be understood by considering the examples. 
    Say we have a data of students who passes and failed. The features being - time spent for studies, time spent for sleep. Now on the basis of these two features we have to predict if a given student will pass or fail. So we have binary outputs, labelled with 'pass' and 'fail'. This kind of problem is known as classification problem. 
    Now let us try to understand the other one. Let us take the example of house sell price. Say, this depends on the area of the house, age and locality. Now for a given house for a given area, locality and age we need to predict the price. This kind of problem is regression problem.

**6. Briefly describe the SVM model.** 

- Ans : Support Vector Machine or SVM is one of the most popular Supervised Learning algorithms, which is used for Classification as well as Regression problems. However, primarily, it is used for Classification problems in Machine Learning.

    The goal of the SVM algorithm is to create the best line or decision boundary that can segregate n-dimensional space into classes so that we can easily put the new data point in the correct category in the future. This best decision boundary is called a hyperplane.

    

    SVM chooses the extreme points/vectors that help in creating the hyperplane. These extreme cases are called as support vectors, and hence algorithm is termed as Support Vector Machine. **7. In SVM, what is the cost of misclassification?**


- Ans : The cost function for SVM consists to two parts. One is related to the hyperplane and the second part is related to the penalty due to misidentification. To understand this properly, let us take the example of linear SVM. 



Looking at the scatter plot by two features X1, X2 as below. The hard and soft margins are shown with black solid and dashed lines respectively. The black solid is the decision boundary. 

Now it can be easily seen that the hyperplane and the margins can be represented as straight lines of the form   
$\vec{\beta} . \vec{X} + \beta_0$ = 0   
Using this equation, the objective is to find a set of values for the vector such that two hyperplanes, represented by the equations below, can be specified.
    $\vec{\beta} .\vec{X} + \beta_0 \ge$ 1  and $\vec{\beta} .\vec{X} + \beta_0 \le$ - 1 . 
    
This is to ensure that all the data instances that belong to one class falls above one hyperplane and all the data instances belonging to the other class falls below another hyperplane. According to vector geometry, the distance of these planesshould be $\frac{2}{\vec{\beta}}$. It is quite obvious that in order to maximize the
distance between hyperplanes, the value of should be $\vec{\beta}$ minimized. So, in summary, the task of SVM is to solve the optimization problem:  min( $\frac{\vec{\beta}}{2}$).  This is for linear kernel. 

Now for non linear data we introduce a slack variable $\zeta$, which provides some soft margin for data instances in one class that fall on the wrong side of the hyperplane. For particles that are on wrong sides, $\zeta \gt 0$

A cost value C is imposed on all such data instances that fall on the wrong side of the hyperplane. The task of SVM is now to minimize the total cost due to such data instances in order to solve the revised optimization problem:  min($\frac{\vec{\beta . \beta}}{2}$) + C $\sum_i \zeta_i$

**9. In the SVM model, define the kernel.** 

- Ans : SVM has a technique called the kernel trick to deal with non-linearly separable data. These are functions which can transform lower dimensional input space to a higher dimensional space. In the process, it converts linearly non-separable data to a linearly separable data. These functions are called kernels. 

The kernel functions can be of different type. For example linear, nonlinear, polynomial, radial basis function (RBF), and sigmoid.

**16. Explain the decision tree algorithm in a few words.** 

- Ans : A decision tree is a non-parametric supervised learning algorithm, which is utilized for both classification and regression tasks. It has a hierarchical, tree structure, which consists of a root node, branches, internal nodes and leaf nodes. Internal nodes represent the features of a dataset, branches represent the decision rules and each leaf node represents the outcome.

    A decision tree starts with a root node, which does not have any incoming branches. The outgoing branches from the root node then feed into the internal nodes, also known as decision nodes. Based on the available features, both node types conduct evaluations to form homogenous subsets, which are denoted by leaf nodes, or terminal nodes. The leaf nodes represent all the possible outcomes within the dataset. 






**17. What is the difference between a node and a leaf in a decision tree?** 

- Ans : Each node (or decision node) of a decision tree corresponds to one of the feature vector. From every node, there are edges to children, wherein there is an edge for each of the possible values (or range of values) of the feature associated with the node. The tree terminates at different leaf nodes (or terminal nodes) where each leaf node represents a possible value for the output variable.


**18. What is a decision tree&#39;s entropy?** 

- Ans : Entropy can be defined as a measure of the purity of the sub split. Entropy always lies between 0 to 1. The entropy of any split can be calculate by this formula. $E = - \sum_i^n p_i log(p_i)$,    where $n$ is the number of different classes, $p_i$ is the probability of $i^{th}$ class labels. 


**19. In a decision tree, define knowledge gain.**
- Ans : The information gain is the decrease in entropy ($E$) after a data set is split according to a particular attribute (A). This information is very important parameter to decide the root node and the later decision nodes. The features for which the information gain is high are considered as root node. 
  If the information gain is 0, it means that there is no reduction in entropy due to split of the data set according to that particular feature. On the other hand, the maximum amount of information gain which may happen is the entropy of the data set before the split.


**20. Choose three advantages of the decision tree approach and write them down.** 

- Ans : Advantages : 

    - Easy to interpret: The Boolean logic and visual representations of decision trees make them easier to understand and consume. The hierarchical nature of a decision tree also makes it easy to see which attributes are most important, which isnt always clear with other algorithms, like neural networks.

    - Little to no data preparation required: Decision trees have a number of characteristics, which make it more flexible than other classifiers. **21. Make a list of three flaws in the decision tree process.** 

- Ans : Flaws : 

    - Prone to overfitting: Complex decision trees tend to overfit and do not generalize well to new data. This scenario can be avoided through the processes of pre-pruning or post-pruning. Pre-pruning halts tree growth when there is insufficient data while post-pruning removes subtrees with inadequate data after tree construction. 

    - High variance estimators: Small variations within data can produce a very different decision tree. Bagging, or the averaging of estimates, can be a method of reducing variance of decision trees. However, this approach is limited as it can lead to highly correlated predictors.  

    - More costly: Given that decision trees take a greedy search approach during construction, they can be more expensive to train compared to other algorithms. 


**22. Briefly describe the random forest model.**

- Ans : Random forest is a commonly-used machine learning algorithm which combines the output of multiple decision trees to reach a single result. This consists of many decisions trees. It uses bagging and feature randomness when building each individual tree to try to create an uncorrelated forest of trees whose prediction by committee is more accurate than that of any individual tree.
