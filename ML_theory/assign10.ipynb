**1. Define the Bayesian interpretation of probability.** 

- Ans : Bayesian interpretation of probability is a non-frequentist approach. In this case, instead of performing infinite trials to obtain the probability of an event to occur, one tries to estimate the uncertainty of an event to occur given a prior knowledge/belief. The most important thing in Bayesian approach is, as the belief changes the probability changes. 


**2. Define probability of a union of two events with equation.** 

- Ans : Let $p(A)$ is the probability of event A to occur and $p(B)$ is the probability of event B to occur.  

The union of two events A and B is given as $p(A \cup B) = p(A) + p(B) - p(A \cap B)$, $p(A \cap B)$ is the probability of occuring both A and B. If A and B are mutually exclusive then,  $p(A \cup B) = p(A) + p(B)$. 


**3. What is joint probability? What is its formula?** 

- Ans : The probability of joint events of A and B is given as $p(A,B) = p(A \cap B) = p(A|B) p(B)$, where $p(A|B)$ is defined as the conditional probability of event A happening if event B happens. 


**4. What is chain rule of probability?** 

- Ans : From the above conditional probability, we can find $p(A)$ after summing up all the contributions from B. e.g. $p(A) = \sum_B p(A \cap B) = \sum_b p(A|B = b) . p(B=  b)$. In a more generalized way, this can be written as : 

     $p(A_1 \cap A_2 \cap A_3 \cap ........ A_N) = p(A_1) . p(A_2|A_1) . p(A_3| A_2, A_1) . ....... p(A_N|A_{N-1} A_{N-2}....A_1) $

    This is known as chain rule of probability.


**5. What is conditional probability means? What is the formula of it?** 

- Ans : Conditional Probability means, given the probability of an event A to occur, what is the probability of event B to occur. This is expressed as $p(B|A) = p(A \cap B)/p(A)$, p(A) should be greater than 0. 


**6. What are continuous random variables?** 

- Ans : A continuous random variable is one which takes an infinite number of possible values. Continuous random variables are usually measurements. Examples include height, weight, the amount of sugar in an orange, the time required to run a mile.


**7. What are Bernoulli distributions? What is the formula of it?** 

- Ans : When the outcome of a trial is either ‘success’ or ‘failure’, then the behaviour of the random variable X can be represented by Bernoulli distribution. Such trials or experiments are termed as Bernoulli trials.

    If $p$ is the probability of an event to occur and say, we have $k$ is the number of possible outcomes, then Bernoulli's distribution is given as : $f(k;p) = p^k (1-p)^{(1-k)}$, where $k \in {0,1}$.  




**8. What is binomial distribution? What is the formula?** 

- Ans : If n independent Bernoulli trials are performed and X represents the number of success in those n trials, then X is called a binomial random variable. That’s the reason a Bernoulli random variable is a special case of binomial random variable with parameters (1, p). 

    Binomial distribution is expressed as : $f(k,n,p) = Pr(X = k) =   ^nC_k p^k (1-p)^{n-k}$



**9. What is Poisson distribution? What is the formula?**

- Ans : The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event.

    A discrete random variable X is said to have a Poisson distribution, with parameter λ > 0 : 
    $f(k, \lambda) = \frac{\lambda^k e^{\lambda}}{k !}$
    
    It can be shown than $\lambda$  is the mean of the Poisson Distribution. 
    
    

**10. Define covariance.** 

- Ans :  The covariance between two random variables X and Y measure the degree to which X and Y are (linearly) related, which means how X varies with Y and vice versa. Covariance is given as : 

    $Cov(X,Y) = E[(X - E[X])(Y - E[Y])] = E[X,Y] - E[X] E[Y] = \frac{1}{N-1} \sum_i (x_i - \bar{x}) . (y_i - \bar{y})$

    As the variance is the measure of how a random variable varies with itself, similarly the covariance is the measure of how two random variables vary with each other. Covariance can be between 0 and infinity. 





**11. Define correlation** 

- Ans : Many a times covariance does not provide enough information about relationship among the random variables. Instead normalized covariance does the task. This is known as correlation. Mathematically it is expressed as : 

    $corr(X,Y) = \frac{Cov(X,Y)}{\sqrt{Var(X).Var(Y}}$ 




**12. Define sampling with replacement. Give example.** 

- Ans : While choosing the samples from the population if each object chosen is returned to the population before the next object is chosen, then it is called the sampling with replacement. In this case, repetitions are allowed. That means, if the sample size n is chosen from the population size of N, then the number of such samples is $N^N$. 

  As an example, one may think of choosing a sample of 2 balls from a population of 3 balls. And repetations are allowed. In that case, each ball can be chosen thrice. Let label the balls as R, G and B. So, we can have (R,R), (R,B), (R,G), (G,R), (G,G), (G,B), (B,B), (B,R), (B,G). So we have total 9 combinations.  This same thing can also be understood as, we need a sample of 2 balls from a population of 3 balls. As repetations are allowed, the first ball of the sample can be chosen in 3 ways and the second ball of the sample can also be chosen in 3 ways. i.e. 3 x 3 = 3$^2$ = 9.  



**13. What is sampling without replacement? Give example.** 

- Ans : If we don’t return the object being chosen to the population before choosing the next object, then the random sample of size n is defined as the unordered subset of n objects from the population and called sampling without replacement. The number of such samples that can be drawn from the population size of N is : 
$^NC_n = \frac{N!}{N! (N-n)!}$

    From the above example, 2 balls from 3 balls (without repetation) can be chosen in $\frac{3!}{2! 1!}$ = 3 ways. 


**14. What is hypothesis? Give example.**

- Ans : While dealing with random variables a common situation is when we have to make certain decisions or choices based on the observations or data which are random in nature. The solutions for dealing with these situations is called decision theory or hypothesis testing and it is a widely used process in real life situations. The key component of machine learning is to use a sample- based training data which can be used to represent the larger set of actual data and it is important to estimate how confidently an outcome can be related to the behaviour of the training data so that the decisions on the actual data can be made. So, hypothesis testing is an integral part of machine learning.  

    In terms of statistics, a hypothesis is an assumption about the probability law of the random variables. Our task is to verify if the assumption is correct or not. 
    
    To understand this, let us take a very simple example. Let us consider a population of N students and the assumption is the mean height is h1 cm. Now, along with this the mean height and standard deviation of the height for the population is given (say). 
    
    Now here we are assuming that the mean height is h1. This is known as null hypothesis, $H_0$. The other assumption can be that, the mean height is not h1. This is known as alternative hypothesis, denoted as $H_1$.  
    Given this, using different statistical tests, we need to confirm that which statement/assumption is true. This is known as Hypothesis testing.
